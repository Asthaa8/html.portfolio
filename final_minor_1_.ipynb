```python
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
# Python code to read image
import cv2
p = cv2.imread("minor image.jpg",0)
cv2_imshow(p)
```

```python
hist = cv2.calcHist([p],[0],None,[256],[0,256])
```

```python
from matplotlib import pyplot as plt
plt.plot(hist)
plt.show()
```

```python
#normalise the histogram
hist /= hist.sum()
plt.plot(hist)
plt.show()
```

```python
# Otsu thresholding is applied
ret,thresh1 = cv2.threshold(p, 0, 255, cv2.THRESH_BINARY +cv2.THRESH_OTSU)
```

```python
# Otsu threhold is obtained
ret
```

```python
# Global thresholding using Otsu threshold value
ret,thresh2 = cv2.threshold(p, 149, 255, cv2.THRESH_BINARY )
cv2_imshow(p)
```

```python
hist = cv2.calcHist([thresh2],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
#Otsu thresholding after gaussian filtering
filter= cv2.GaussianBlur(p,(5,5), 0)
ret3,thresh3 = cv2.threshold(filter,0, 255, cv2.THRESH_BINARY+ cv2.THRESH_OTSU )
cv2_imshow(filter)
```

```python
hist = cv2.calcHist([thresh3],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
#Otsu thresholding using arithmetic mean
thresh4 = cv2.adaptiveThreshold(filter, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,21,10)
cv2_imshow(thresh4)
```

```python
hist = cv2.calcHist([thresh4],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
#Otsu thresholding using gaussian average
thresh5 = cv2.adaptiveThreshold(filter, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,21,5)
cv2_imshow(thresh5)
```

```python
hist = cv2.calcHist([thresh5],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```
